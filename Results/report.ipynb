```python
import seaborn as sns
import matplotlib.pyplot as plt

import pandas as pd
```

```python
def printstat(A, B):
    A = round(A,4)
    B = round(B,4)
    
    wins = sum(A > B)
    losses = sum(A < B)
    ties = sum(A == B)
    
    print("Wins/Loss/Ties:\t{}\t{}\t{}".format(wins, losses, ties))
```

```python
R = pd.read_csv("ressum5.csv", sep=",")

#first argument wins against the second argument
printstat(R["VEGNN"], R["GNN"])
printstat(R["VEGNNHS100"], R["VEGNN"])
printstat(R["VEGNNHS500"], R["VEGNN"])
printstat(R["VEGNNHS1000"], R["VEGNN"])
```

```python
VG = pd.DataFrame(R, columns = ["DATASET", "GNN", "VEGNN"]).sort_values(by=["VEGNN"], ascending=True)
VG.plot.bar(x="DATASET", figsize=(20,10))
```

```python
"""
Horizontal bar plots for comparison
"""
Rs = R.sort_values(by=["VEGNN"], ascending=False)

sns.set()

# Initialize the matplotlib figure
f, ax = plt.subplots(figsize=(9, 20))

#sns.set_color_codes("pastel")
sns.set_color_codes("dark")

sns.barplot(x="VEGNN", y="DATASET", data=Rs,
            label="VEGNN", color="r")

sns.barplot(x="GNN", y="DATASET", data=Rs,
            label="GNN", color="g")

# Add a legend and informative axis label
ax.legend(ncol=2, loc="lower right", frameon=True)
ax.set(ylabel="",
       xlabel="Performance of VEGNN against GNN")
sns.despine(left=True, bottom=True)
```
